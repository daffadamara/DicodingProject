### **Exploratory Data Analysis**

Pertanyaan:
1. Apakah ada korelasi antara berbagai polutan udara (SO2, NO2, CO, O3)?
2. Bagaimana konsentrasi polutan udara di berbagai lokasi stasiun?
3. Apakah ada tren atau pola yang terlihat pada tingkat polutan sepanjang tahun?
4. Pada stasiun mana suhu mencapai derajat terendah dan tertingginya?
5. Pada stasiun mana curah hujan mencapai volume tertingginya?

Sebelum melanjutkan proses analisis, data frame perlu di merge agar menjadi satu kesatuan.

#### **Merging Data Frame**

```python
df = [Data_Aotizhongxin, Data_Changping, Data_Dingling, Data_Dongsi, Data_Guanyuan, Data_Gucheng, Data_Huairou, Data_Nongzhanguan, Data_Shunyi, Data_Tiantan, Data_Wanliu, Data_Wanshouxigong]
air_quality_merge = pd.concat(df)

air_quality_merge.head()
```

#### **Exploring**

```python
air_quality_merge.describe(include='all')
```

Dari informasi di atas kita mendapatkan bahwa:
- Terdapat 12 stasiun yang dianalisis
- Suhu tertinggi berada di kisaran 41,6 derajat celsius
- Curah hujan tertinggi berada di kisaran 72,5 mm
- Polutan dengan kadar tertinggi adalah CO (Karbon Monoksida)
- Observasi dimulai sejak tahun 2013 sampai dengan 2017

**Membuat Pivot Table**

Temperatur dan Curah Hujan di Berbagai Stasiun

```python
air_quality_merge.groupby(by="station").agg({
    "TEMP": ["max", "min"],
    "RAIN": ["max", "min"],
})
```

Hasil analisis:
- Suhu tertinggi terjadi pada stasiun Gucheng sedangkan suhu terendah terjadi pada stasiun Wanliu.
- Curah hujan tertinggi terdapat pada stasiun Aotizhongxin dan Wanliu.

Rata-Rata Konsentrasi Polutan di Berbagai Stasiun

```python
air_quality_merge.groupby(by="station").agg({
    "SO2": ["mean"],
    "NO2": ["mean"],
    "CO": ["mean"],
    "O3": ["mean"]
})
```

Hasil analisis:
- Rata-rata kadar SO2 tertinggi terdapat pada stasiun Nongzhanguan
- Rata-rata kadar NO2 tertinggi terdapat pada stasiun Wanliu
- Rata-rata kadar CO tertinggi terdapat pada stasiun Wanshouxigong
- Rata-rata kadar O3 tertinggi terdapat pada stasiun Huairou

```python
pollutant_mean = air_quality_merge.groupby(by="station").agg({
    "SO2": ["mean"],
    "NO2": ["mean"],
    "CO": ["mean"],
    "O3": ["mean"]
})

pollutant_mean['kadar_polusi'] = pollutant_mean.mean(axis=1)

print(pollutant_mean)
```

Hasil analisis:
- Stasiun Wanshouxigong memiliki konsentrasi polusi tertinggi dengan rata-rata polutan sebesar 374,83
- Stasiun Dingling memiliki konsentrasi polusi terendah dengan rata-rata polutan sebesar 253,19

#### **Plotting**

Untuk menjawab pertanyaan "Apakah ada korelasi antara berbagai polutan udara (SO2, NO2, CO, O3)?" diperlukan analisis lebih lanjut dan plot berupa scatter plot untuk mengetahui persebarannya.

```python
air_quality_merge.corr(method= 'spearman', numeric_only=True)
```

Tabel di atas merupakan tabel korelasi antar variabel. Perlu dibuat plot agar lebih mudah dianalisis.

```python
correlation_matrix = air_quality_merge[['SO2', 'NO2', 'CO', 'O3']].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f") # Matrix heatmap
plt.title('Correlation Matrix of SO2, NO2, CO, and O3')
plt.show()
```

Dari gambar di atas dapat diketahui bahwa:
- SO2 memiliki korelasi yang cukup kuat terhadap NO2 (0,49) dan CO (0,53) dan berkorelasi positif.
- NO2 memiliki korelasi yang kuat terhadap CO (0,69) dan berkorelasi positif.
- O3 memiliki korelasi yang cukup kuat terhadap NO2 (-0,46) dan cukup lemah terhadap CO (-0,31) dan berkorelasi negatif.
- O3 memiliki korelasi yang sangat lemah terhadap SO2 (-0,16) dan berkorelasi negatif.

```python
air_quality_merge['Year'] = air_quality_merge['Datetime'].dt.year


# Group the data by year and calculate the mean level of each pollutant
pollutant_means_over_years = air_quality_merge.groupby('Year')[['SO2', 'NO2', 'O3']].mean()


# Create the plot
plt.figure(figsize=(10, 6))

# Plot SO2 levels
plt.plot(pollutant_means_over_years.index, pollutant_means_over_years['SO2'], label='SO2')

# Plot NO2 levels
plt.plot(pollutant_means_over_years.index, pollutant_means_over_years['NO2'], label='NO2')

# Plot O3 levels
plt.plot(pollutant_means_over_years.index, pollutant_means_over_years['O3'], label='O3')

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Pollutant Level')
plt.title('Trends of Air Pollutants Over the Years')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

Dari gambar di atas dapat diketahui bahwa:
- Tidak terdapat pola tren musiman yang terjadi disepanjang tahun sampai tahun 2017.
- Kadar O3 dan NO2 memiliki kemiripan pola dan tren disepanjang tahun sampai tahun 2017.

##### **Plot-Plot Lainnya**

```python
stations = pollutant_mean.index
mean_pollutant_levels = pollutant_mean['kadar_polusi']

plt.figure(figsize=(10, 6))
plt.bar(stations, mean_pollutant_levels, color='skyblue')

plt.xlabel('Station Name')
plt.ylabel('Mean Pollutant Level')
plt.title('Mean Pollutant Levels per Station')
plt.xticks(rotation=45, ha='right') 

plt.tight_layout()
plt.show()
```

```python
volume_mean = air_quality_merge.groupby(by="station").agg({
    "TEMP": ["max", "min"],
    "RAIN": ["max", "min"],
})

stations = volume_mean.index
max_temp_levels = volume_mean['TEMP']['max']
min_temp_levels = volume_mean['TEMP']['min']

bar_width = 0.35

r1 = range(len(stations))
r2 = [x + bar_width for x in r1]

plt.figure(figsize=(12, 6))
plt.bar(r1, max_temp_levels, color='skyblue', width=bar_width, edgecolor='grey', label='Max Temperature')
plt.bar(r2, min_temp_levels, color='lightgreen', width=bar_width, edgecolor='grey', label='Min Temperature')

plt.xlabel('Station Name', fontsize=14)
plt.ylabel('Temperature Level', fontsize=14)
plt.title('Max and Min Temperature Levels per Station', fontsize=16)
plt.xticks([r + bar_width/2 for r in range(len(stations))], stations, rotation=45, ha='right')

plt.legend()

plt.tight_layout()
plt.show()
```
